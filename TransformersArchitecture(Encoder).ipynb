**Encoder Architecture of Transformers:**

We have one part as encoder and another part as decoder.In original "Attention is all you need" , they used 6 blocks in encoder block and 6 blocks in decoder block.All the 6 blocks in encoder and 6 blocks in decoder are similar,so we just have to understand the one block in encoder and one block in decoder as shown below.

1.jpg

Consider one encoder block among those 6 blocks. Each encoder block in consist of feed-forward-neural network and self_attention block. We have "add" which represents the residual network and "norm" which represents layer normalization block.

1.jpg

So we have represented the encoder block with only one feed forward neural network and one self attention block. But actually we have such 6 blocks which contains the ff-neural network and self attention block. Input is provided to the 1st block , the output of this 1st block is provided to the 2nd block , output of the 2nd block is provided to the 3rd block and so on. And at the end , the output of the 6th block is provided to the decoder block.

**How Encoder Bloack Works?**

Let we have a example sentence of "How are you?".In reality we send batch of sentence but for example consider we are providing this one sentence only. We do tokenization of the input sentence which means sentence is divided into words.Then these each word is passed to the embeddings layer which will convert the words into vector representation of each word. For adding positional information of the word,we do positional encoding of the vector representation of each word.Now we will get x1 , x2 and x3 as positional encoding embedding vectors of words How,Are and You respectively.

When 1st vector is passed to the multihead attention block , it creates z1 vector which understands the contextual word embeddings of the word.We have Add and Norm block in encoder block.We exxecute the layer normalization process to the contextual word embeddings of the input word. We will do this for all the 3 words.We perform the normalization for stabalization of traning process and fasten the training process.Now we have z1norm z2norm and z3norm which is the normalized contextual word embeddings of the input words How , are and you respectively.

These three vectors will go to the feed forward network.We will stack these 3 vectors to create a 3 * 512 size matrix.There are 2 layers in ff neural network.Input to the ff-neural network is 512 size. Activation function is relu in 1st layer and linear in 2nd layer.The output that we got from ff-neural netword is also 3 * 512 size matrix.With the help of relu activation function , we could add the non-linearity in the model. There is Add block in encoder architecture and that is for the residual network connection where we are passing normalized word embeddings of the input word to the output of ff-neural network. All the vectors are of 512 size and we will perform addition operation and we will get the y1' , y2' and y3' which is again a size of 512.We get the new set of vectors to which we have to apply the layer normalization step.The output of layer normalization step will give us y1norm , y2norm and y3norm.

These y1norm , y2norm and y3norm will go the next encoder block.These three vectors will work as the input to the next encoder block.All the weights and biases in 1st encoder block will be different from the all other 5 encoder blocks.Each six bloacks of the encoder will give the output in size 512 size.

* 1) Why to use residual connection?
* Ans - In the original research paper of "Attention is all you need" , there is no clear cut explanation given as to why did we used residual connection here.We just know that these connections are important.The reason for using residual network could be to stabalize the training when we have a deep neural network.By residual network , we provide a alternate path for gradient flow and as a result gradients doesnt shirnk fast and it keeps its value.

* 2) Why to use a feed forward neural nework?
* Ans- If we use FFNN , we can catch the non-linearity present in the data.This is also the reason why we used Relu activation function.

* 3) Why to use 6 encoder blocks?
* Ans- Human language which is going to be the input to the transformer is very complex to understand.If we use only 1 encoder block then we get bad accuracy or results.So the people thought of using 6 encoder blocks in the transformer encoder architecture.

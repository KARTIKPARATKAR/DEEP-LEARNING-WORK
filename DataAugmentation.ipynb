**Data Augmentation-**

Data augmentation is a technique using which we reduce overfitting and create new data if you are short of the data for training the model. Let you are working on image data and number of images are very less then you can create new similar images and train the model on sufficient amount of data.We use imagedatagenerator in keras.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat.jpeg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img) #This is not the numpy array. It is PIL image
```

Main Code for Data Augmentation-

```python
datagen = ImageDataGenerator(
        rotation_range=40,  #Angle to which you want to rotate the image
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
#COnverting image into numpy array
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
#We have to tell how many images we will give.
#For now we are giving only one image.
#So we will define this 1 image as follows:
input_batch = img.reshape(1,200,200,3)
```

```python
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='/home',save_prefix='cat',save_format='jpeg'):
  i+=1
  if i>20:
    break
#After running this code, 20 images will be created and stored in home folder of google collab.
```

Go to the Home folder , you will find 20 cat images created with the help data augmentation technique.

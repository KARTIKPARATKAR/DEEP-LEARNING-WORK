**What is Normalization?**

It is a process of transforming data or model outputs to have specific statistical properties typically a mean of zero and a variance of one.
We can apply normalization on input layer as well as in hidden layers of neural network.

Normalization-

* improves training stability.
* Faster Convergence
* Reduces the internal covariate shift
* Introduces a regularization effects like reducing the overfitting.

There are two ways by which we can do activation normalization are:

1)Batch Normalization

2)Layer Normalization

**Why we use Layer Normalization In Transformer Architecture and not Batch Normalization?**

Batch normalization doesnt work efficiently with self_attention. Or efficiently we can say that batch normalization cant work properly on text data.

Let we have a dataset where we have different sentences with different number of words present in it.let we have a longest sentence with 30 words and smallest sentence with 5 words. To be consistent , we will have to do zero padding in the smallest sentence to increase the number of words equal to longest number of words.We will have lot of unnecessary zeroes in the calculation. And when we will go for calculation of standard deviation and variance we will have to unnecessarily consider zeroes in our calculations.And this is the reason why we cant use batch normalization in trsnaformer and self_attention.

**Layer Normalization in Transformers:**

When we aply batch normalization , we normalize across the batch . And we apply layer normalization , then we normalize across the feature which means in row wise direction.Layer Normalization operates independently on each data point.

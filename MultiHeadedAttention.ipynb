**Recap Of Self-Attention-**

It is a technique used to create a contextual word embeddings of the input text data.One hot encoding,bag-of-words and embeddings technique does not give us the contextual vector representation of the text data.But using self-attention , we can create a contextual embedding according to the data used and task we are performing.Contextual word embeddings means  that when we have two sentences "money bank" and "river band" then here the meaning of word bank is different in both the sentences so the numerical or vector representation of both the words should also be different and self attention has the power to represent same word with different numers or vectors which is known as contextual word embeddings.

**Problem with Self-Attention-**

Self attention can only understand the single perspective. If we have multiple perspective in a data so self-attention is not able to capture both the perspective.

Let we have a sentence like "The man saw the astronomer with a telescope". This sentence can have 2 meanings like "Aadmi ne astronomer ko telescope se dekha" and "Aadmi ne telescope ke saath astronomer ko dekha". So we have a ambiguity in this sentence as per as meaning of the sentence is concerned.So when we apply self attention to this sentence then only one perspective or only one meaning is being captured. The other meaning or perspective is rejected because self attention is not able to capture the both scenario or meaning of the sentence.

To solve this issue , we have a multi-head attention in transformer.

###**MultiHead Attention -**

So when we have ambiguity in meaning of the sentence and that sentence has multiple meaning. When we aply self attention to this sentence , self-attention technique can only capture one meaning or one context. So Multi-Head attention suggest that what if we use 2 self-attention modules.Then we can capture both the meaning or context of the input sentence. So since we are using 2 self-attention modules we cll it as a multi-head attention.

**Simple self-attention mechanism-**

We have a input sentence "money bank".We create a word embeddings of both the words.Then we create a matrix for query , key and value part with randomly initialized value which will update its value during training of the model.These 3 matrices will be multiplied by word embeddings of each word which gives us the query , key and value vector for calculation of self-attention.

**MultiHead Self-Attention-**

The difference between simple self attention and multihead self attention is that we create a 2 set of matrix for key , query and value part. So when we multiply the word - embeddings of the words with these 2 set of matrices , we get 2 set of vectors for query , key and value.So as we have two set of vectors query , key and value we can generate the 2 self-attentions contexual word embeddings of the each word in a sentence.

So here we have multihead attention block which consist of two self attention blocks.We are passing embeddings of both the input words money and bank to the multihead attention module in both self attention 1 and self attention 2. Each self attention module is creating the contextual self embeddings of both the words which means we have 2 contextual word embeddings of both money and bank.As we have created two contextual word-embeddings , we can capture the two different meanings of the each word present in the sentence.We can create 3 head or 4 head or whatever number of contextual word embeddings according to our dataset and the task we are performing.

**Long Short Term Memory [LSTM] -**

We want to maintain two types of context in LSTM is that , Long Term Memory(LTM) and Short Term Memory(STM). LSTM establishes the communication between LTM and STM.

We get three inputs to the LSTM , previous cell state , previous hidden state and input for current time t1.

We get two output in LSTM , current hidden state and current cell state.

Inside the architecture of LSTM we process two things, updation of cell state and calculation of cell state.We also remove the things based on current input and add the things based on current input in long term memory(LTM).

 **Gates In LSTM**

There are 3 gates in LSTM ,

Forget Gate=To remove something from cell state/from long term memory

Input Gate=To add something to the cell state/to long term memory

Output Gate=To calculate hidden state for next time state.

Ct and Ht are two vectore with same dimensions. If Ct has 3 numbers in it then Ht will also have 3 numbers in it.

Xt is also the vector.Xt is one word converted to the vector.We pass every word to the LSTM which is nothing but the Xt.

ft , it , ctbar and ot are also the vectors with same size as Ct and Ht.ft is present in forget gate , it and ctbar are present in input gate and ot is present in output gate.

**Decoder Architecture in Prediction/Inference time-**

Let we have a machine translation task of conversion of english sentence to hinndi sentence.We have trained the model on available dataset and now ready to translate unknown english sentence into hindi sentence. We have unknown english sentence "We are friends".

Encoder behaviour works similar in training time and prediction time.Decoder behaves differently in training time and prediction time.Decoder works autoregressive during the time of training and decoder works non-autoregressive during the time of prediction.Autoregressive models means prediction is dependant on the output of previous time steps.

The unknown sentence "We are freiends" is given as the input to the encoder block architecture which creates a contextual word embedding vector for each word in sentence.Now decoder work start who has to predict the hindi conversion of the input sentence.

To start the decoder process , we pass a start sos token which tells decoder to start giving hindi conversion.sos is an english word and model cant understand this sos word so we convert the sos token into vectors with the help of embedding layer.Then we apply positional encoding vector and get the output as x1.In decoder block , we have masked attention block , cross attention block and feed forward neural network block. x1 is fed to masked multihead attention.We have 3 matrix of wq,wk and wv which when multiplied with x1 , we get query , key and value vectors. We perform dot product of query and key vector which gives a scalar value. We perform scaling on this value and perform the softmax function which gives the  attention score.This attention score tells us that how the start of sentence token is similar to itself.This similarity is then multiplied with the value vector which is nothing but a z1 vector.And this completes the masked-self-attention process.

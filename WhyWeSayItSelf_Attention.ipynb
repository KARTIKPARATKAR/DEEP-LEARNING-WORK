**How self attention is called attention mechanism?**

 We were trying to solve the problem of language translation like from enlish to hindi conversion. We used encoder - decoder architecture to execute this task. Encoder and decoder has LSTM inside it.Ecoder block compress the input information and maintains the context vector which passes the information to the decoder for final output.Context vector cant compress the information when no of input words in sentence increses to more than 30 and it reduces the quality of translation.

 To solve this problem we have attention mechanism in transformers.

**How self_attention is similar to Luoung Attention?**

When we learned self-attention that time we didnt have any encoder or decoder architecture. We just had sentence and we had to create a contextual word embeddings of each word in a sentence.We created 3 vectors for each word for value,key and query part of the word embedding which will be helpful for calculation of contextual word embeddings of that word.

Query part in self attention is analogous to decoder hidden states in Luoung attention.

Key part and value part in self attention is analogous to encoder hidden states in Luoung attention.

Because of above analogy of self attention and Luoung attention ,we call Self attention as the attention mechanism.

**Sequence Tasks and Its Types:**

RNN is used when we have sequential data like text in any language.

3 types of RNN -

1) many to one RNN like sentiment analysis where we have movie review as input and output as positive or negative.

2) one to many RNN like in input there no sequence and sequence is there in output like input is image and we give information in text about that image.

3) many to many RNN where we have sequence in input as well in output.

 We have synchronous many to many RNN where input and output length is same and example is parts of speech tagging where we give a x number of input numbers and we get x number of output parts of speech.

Input:  ["I", "am", "happy"]

Output: ["PRON", "VERB", "ADJ"]


We have asynchronous many to many RNN where input and output length is not same and example would be machine translation where we give text in one language with x number of words in it and get the output in another language with y number of words in it.

**Stage 1 - Encoder Decoder Architecture -**

There was a guy called Iilya Sutskever from Google who published research paper with tile "Sequence to Sequence Learning wit Neural Networks" which got huge popularity in Encoder Decoder architecture domain.

IN seq2seq task,we have a sequence as input and we want to convert it into another sequence in output.In Encoder Decoder architecture , incoder will process the information word by word at every time stamps and compress this information.This compressed information is passed to the decoder and decoder gives the final output.This ENcoder-Decoder used to work better when we gave small sentences.But when we gave large paragraphs , it was not giving the better results becuase we are compressing the informaton and it is upto context vector to keep all the informstion from encoder and pass it to the decoder. When we have large number of words this context vector doesnt work properly.

1.jpg

**Stage 2 - Attention Mechanism**

Encoder Decoder architecture was not able to give better result when we had bigger paragraphs. TO solve this issue we introduced Attension Mechanism.

There is a guy called Yoshua Benjio and his team who published research paper with title "Neural Network Translation By Jointly Learning To Align And Translate" which introduced Attention Mechanism in Seq2Seq Encoder Decoder Attention Architecture.

In attention based encoder - decoder architecture,there is no single context vector,at any step of the decoder , decoder has all the information of internal step of encoder.Attention is basically a mechanism which finds out the state in encoder which is useful in giving output word.

1.jpg

**Stage 3 - Transformers-**

Because we are calculating the separate context vector for each time step of the decoder the computational complexity/training time was increasing.Let we have n words in input and m words in output and for even calculating the one output word from m , we have to calculate the m*n context vector which increases the complexity and training time.To solve this problem , transformer architecture was introduced.

The research paper with title "Attention Is All You Need" which suggested there is no need of RNN or LSTM or GRU layer in encoder and decoder blocks. Only attention mechanism is enough.Self attention was implemented in both encoder and decoder blocks.In transformers, they can see all the words in one go which decrease the training time of the model.

**Stage 4 - Transfer Learning -**

Training the transformer architecture model from scratch is very difficult because it requires compatible hardware,more time of training and requires lot of data.

We dont have have huge amount of data so we are not able to train the transformer based Seq2Seq model from scratch.This became the huge problem for using transformer.To solve this issue we introduced Transfer Learning .

The research paper with title "Universal Language Model Fine-Tunning For Text Classification" by Jeremy Howard and team suggested that the concept of transfer learning can be used in NLP tasks.Before this paper , Transfer Learning was only used in Machine Vision Tasks.

Till this paper was published , transfer learning was not used in NLP tasks because we felt that one trained model can not do the mutiple NLP taks like machine translation , sentiment analysis , part of speech tagging and etc.

Language modelling is the NLP task in which we train the model to predict the next word.This research paper used Language Modelling as Pretraining.

**How Language modelling as Pretaining was successful?**

1)Rich Feature Learning - Learning to predict the next word is rich in feature.The model learns lot of things while predicting the next word. Some better models derives common sense along with situations.This means that the next word predictor model has lot of knowlege.This knowlege can be used to define a text classification task,question-answe system,text summarization , NER(Named Entity Relations) and Part of Speech task.

2)Huge Availability of Data - If we are using language modelling as Pretraining then we dont need labelled data for training of our transformer architecture.

**Setup for ULMFit(Universal Language Model Fit)-**

The researchers used LSTM variant called AWD LSTM on a dataset of wikipedia text and performed unsupervised pretraining model to predict the next word prediction.When this model was trained,they changed the output and replaced with classifier and gave different dataset like MDB , YELP and etc dataset. The model that we got was tested on different model and output was found state of the art. Thus we conclude that language modelling as pretraining is poweful and gives better result.

**Stage5 LLM's (Large Language Models)-**

Two transformer based language models was introduced by Google is BERT and by OpenAI is GPT(Generative Pre-Trained Model). BERT was Encoder only model and GPT was decoder only model.These bth models was trained on huge dataset which was able to do the text summarization task,question answer task,named entity relation task , part of speech task and sentiment analysis task.

After these two models,anyne could download these models , train these models with limited dataset and after fine tunnig , anyone could get state of the art result.

OpenAI introduced the successive models of GPT in GPT2 and GPT3. These successive models was so big that is was trained on huge dataset along with having huge number of parameters,so people started calling it as Large Language Models.

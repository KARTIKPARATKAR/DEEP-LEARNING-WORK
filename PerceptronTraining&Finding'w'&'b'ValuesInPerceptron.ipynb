**Perceptron Trick-**

Percptron trick is used to adjust the weights of a percptron when it makes a mistake during training.It ensures that the percptron gradually learns a decision boundrary tht correctly separates the data.

Let we have a linear data and we want to separate it with the help of percptron. Perceptron is nothing but a line having equation -- Ax+By+C = 0

So plotting a proper line we need to know the values of varibles A,B and C .

FOr that we will draw a random line with A=1,B=1 and C=0 and check this line correctly distinguishes all the points or not. If not then we will make appropriate changes in values of A B and C so as to distinguish all the points in a dataset is distinguished in two regions.

Here after applying multiple percptron trick,we get a line equation which is nothing but a percptron equation that can classify the two classes separately.



**Writing Percptron Trick Algorithm-**

1) Let equation of perceptron is Ax+Bx+C =0

Here consider A and B as weights and C as bias.

A and B can be considered as weight values C as bias value.
We assign random value to A , B and C and draw a line.

2) This line might not be dividing all the datapoints correctly in two regions.So we will pick a random datapoint and check whether that datapoint is being distinguished correctly or not.

3)If randomly selected datapoint is being distinguished correctly,then we wont change the values of A,B and C which we assigned randomly.

4)If randomly selected datapoint is not being distinguished corectly , then we will make the appropriate changes to the values of A,B and C so as to distinguish this datapoint correctly.

5) We will do this process 1000 times which we call as epochs. i.e. Epoch = 1000. So after selecting 1000 datapoints randomly and checking whether that datapoint is being distinguished correctly or not , we will get the correct values of A, B and C which distinguishes all the datapoints correctly.

6) And as a result we get the line equation which is nothing but a Perceptron equation which we were aiming to distinguish all the datapoints in either of the two classes.

**Convolutional Neural Network(CNN)-**

 CNN is a type of neural network which is specially used when you have grid like data.Grid like data means images(2D grid) which is represented in pixels , videos(3d grid) which is represented as height , width and time and time-series dat which is 1d or 2d grid.

 As there is convolution in name itself, there is a convolution layer present in CNN.In ANN , we use metric multiplication and in CNN , we use convolution operation.


 Layers in CNN-

 * Convolution layer

 * Pooling Layer

 * Fully COnnected Layer


 CNN is inspired by visual cortex of human beings.CNN is most popular neural network architecture in deep learning.



**Why not to use ANN on image data?**


1.jpg

1) Images are the 2d grid of pixels.If we use ANN on image data,we will require high computational cost.Let consider an image of 40 * 40 pixels. If we use ANN then we will have to give 40 inputs to the ANN network,there will be 40 nodes in first hidden layer. So we will have 40 * 40 * 100 =160000 parameters to train in first hidden layer.And mind that this 40*40 pixel of image is very small pixels.We will definately have more numbers of pixels in the image.So as number of pixels increases then the numbers of parameters to train will also increase and it is not feasible to compute for neural network model.


2) As there is lot many connections in ANN if there is more numbers of input,we might face the overfitting problem because we will have all the layers fully connected and it will train good on train data so the model will give good results on train data as it wont give good results on test data.

3) If we use ANN in image data then we might lose some important information like spatial arrangement of pixels. The distance between two pixels is important feature to consider,but when we use ANN , we give 1st row pixels as input resulting in not consideration of distance between two pixels . And so model wont learn the distance of two pixels.

**How CNN is Inspired from Human Visual Cortex-**

CNN architecture is highly inspired from human visual cortex.We have two types of cells one is simple cell and second is complex cell. Simple cell is the orientation cell and feature detector cell also.Simple cell is smaller receptive field and can process smaller portion of area.Complex cell is bigger cell and has bigger receptive field which means complex cell can process bigger portion of area. Simple cell has a work of detection of edge. You pick up any image,eventually images are made up of edges only.So brain divides any image into smaller/minute edges and then create a bigger edge and form the original image to read by the brain.

Consider example of circle,this circle can be divided into n number of edges by simple cell,then after merging these smaller edges we make bigger edge like hexagon , then pentagon , square and eventually a circle.So merging these edges means merging fatures of the image.

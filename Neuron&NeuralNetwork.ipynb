**Neuron-**

It is a fundamental computations unit within an artificial neural network(ANN).

Neurons-
1)Takes Input - Receives multiple inputs which can be raw data or outputs from other neurons.

2)Applies weights and bias -Each input has an associated weight that determines its importance and a bias is added to shift the result.

3)Computes Weighted Sum - The neuron sums up the weighted inputs along with the bias.

4)Applies Activation Function- The weighted sum is passed through an activation function like ReLU, Sigmoid or Tanh.

Types of Neurons-
Input Neurons - Receives raw data.

Hidden Neurons - Intermediate layers that process data to extract patterns.

Output Neurons - Provide the final prediction or classification

**Neural Networks:**

1)Multi-Layer Perceptron(MLP)-

 Used in supervised machine learning problem like regression and classification.It is called as multilayer because it contains an input layer,one-or-more hidden layers and an output layer.

 1.jpg

2)Convolutional Neural Network(CNN)-

  Used in image and video processing problems.
  
  Convolutional layers apply convolutional operations to input images using filters (Kernels) to detect features such as edges , textures and more complex patterns.

  Pooling layers- Downsample the spatial dimensions of the input.

  Activation Functions- Introduces non-linearity to the model allowing it to learn more complex relationships in the data.

  1.jpg



3)Recurrent Neural Network(RNN)-

  In somewhere between hidden layer,we have a feedback mechanism where output on a perticular hidden layer is feed to the previous hidden layer so as to increase the prediction. RNN is used in Natural Language Processing tasks.
  RNN remember past information by feeding the output from one-step into next step.


1.jpg

5)Long Short Term Memory(LSTM)-

  LSTM is a advanced version of RNN  which capture long-term dependencies in sequential data making them ideal for tasks like language translation,speech recognition and time-series forcasting.

  1.jpg
  

4) Auto Encoders-
  
  Let we have a input data and we want to compress it without loosing the quality. We have same number of input layers and output layers but different numbers of hidden layers.

  Encoder compresses the input into a compact representation and capture the most relevent features.

  Decoder reconstructs the input data from this compressed form to make it as similar as possible to the original input.

6)Generative Adverserial Network(GAN's)-

  We can generate audio,video and images from GAN's. GAN's are a class of neural networks that autonomouly learns patterns in the input data to generate new examples resembling the original network.

  Two types of GAN archtecture are -

  1)Generator-- Creates data from random noise which can not be distinguished whether it is real or fake.

  2)Discriminator-- It acts as a critic and tells the received data is real or fake.

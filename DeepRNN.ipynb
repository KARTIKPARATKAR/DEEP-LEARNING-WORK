**Stacked RNN**

Deep RNN means nothing but stacking one  RNN cell on top of each other and then unfolding all of them over time.

1.jpg

Deep RNN/Stacked RNN is used to understand the deep patterns in data.Stacked RNN is capable of understanding the hierarchy present in the data.When we want to build language translation model using encoder-decoder , there we can use deep RNN's.

* When to use Deep RNNs?
* Speech recognition , Machine Translation,Large Dataset,Has Sufficient Computational Power, when u are not satisfied with single RNN.

**Implementation Of Deep RNN**

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding , SimpleRNN , Dense
```

```python
#Loading the dataset
(x_train , y_train) , (x_test , y_test) = imdb.load_data(num_words = 10000)
```

```python
#We want to have a same length of input , so we are using pad_sequences
x_train = pad_sequences(x_train , maxlen = 100)
x_test = pad_sequences(x_test,maxlen=100)
```

```python
#Model defination
model = Sequential([
    Embedding(10000,32,input_shape=(100,)), #10000 dimentional input with 32 dimentional output
    SimpleRNN(5,return_sequences=True),
    SimpleRNN(5),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

We can define Deep LSTM and Deep GRU as like we define the Deep RNN,Just we have to replace RNN with LSTM and GRU.

Since we are increasing the complexity of the model by stacking the layers , the tendency of overfitting increases.

Training time increases as we will use large datasets in stacked/deep RNN/LSTM/GRU .

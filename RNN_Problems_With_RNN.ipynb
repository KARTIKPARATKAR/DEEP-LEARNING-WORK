**RNN -**

RNN are neural network architecture with is suitable for sequential data like text and time-series. RN are not used that much because RNN suffer from two problem . Instead of RNN , we use LSTM(Long - Short - Term - Memory) architecture.

Problems are -

1)Problem of long term dependency-

2)Unstable Training (Exploding Training) -

These two problems arises due to unstable gradients. We have sequential data in RNN . If the sequence is very long which means the number of stpes is more then the things in older steps RNN is not able to remember.This is a problem of long term dependency.

Let we have a problem of predicting a next word in sentence.If our sentence is of lets say 6 words . We know the 5 words and we have to predict 6th word , then RNN might give good result , but if the sentence is of 20 words and we know 19 words and we have to predict the 20th word , then RNN will give bad results because there are 20 time steps and RNN is not able to remember the long term dependency.


Stagnated training means we would not be able to train the RNN model properly resulting in non-updation of weights and bias by gradient descent algorithm.This happens due to exploding gradient problem.

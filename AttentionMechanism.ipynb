**Need of Attention Mechanism**

Attention mechanism is used in Encoder - Decoder Architecture seq2seq models. Pink block is encoder and green block is decoder and yellow cells anre LSTM's.We give one sentence to encoder , encoder preocess this input sentence step by step and generates summary in form of context vector. This compressed information is passed to the decoder via context vector and decoder gives output of the model.

The problem with this approch is that when we give more number of words in input to the encoder. The context vector is not able to remember evrything that is present in input because there are many words present in input data.Load of transfering the data is huge on context vector when more number of words are given in input.In decoder point of view, when decoder is giving perticular output word,it has the information of all the input words and it is static in form of context vector.It would have been better if we had the authority to look at only one input word that is contributing more to give output of the perticular output word.

**What is Attention Mechanism-**

h0,h1,h2 and h3 are hidden state of LSTM present in encoder.s0,s1,s2 and s3 are hidden state of the LSTM in decoder.y0,y1,y2 and y3 are input to the LSTM of Decoder in each time step.c1,c2,c3 and c4 are attention input which is nothing but the time step of encoder LSTM which tells the Decoder LSTM cell of which LSTM state or states from encoder is important for decoder LSTM to give output.

So in some cases decoder LSTM might require only one encoder LSTM state and in some cases decoder LSTM might require more than one encoder LSTM state.

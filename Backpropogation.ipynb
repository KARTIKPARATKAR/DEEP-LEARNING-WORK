**Backpropogation:**

It is an algorithm which is used to train the simple as well complex neural network.

* What does training a neural network mean?

* Ans- We have weights and biases in a neural network.We are intended to find the values of weights and biases for given data in a neural network so that our model will give the better predictions.So process of finding the best values of all weights and biases is known as Training the model.

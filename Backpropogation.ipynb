**Backpropogation:**

It is an algorithm which is used to train the simple as well complex neural network.

* What does training a neural network mean?

* Ans- We have weights and biases in a neural network.We are intended to find the values of weights and biases for given data in a neural network so that our model will give the better predictions.So process of finding the best values of all weights and biases is known as Training the model.

3) What is Convergence?

Convergence is the stage at which even when after updating the bias and weights values with backpropogation algorithm,we are not getting any better accuracy or our loss function is not getting decrease.Then we say that we have reached to Convergence and we will stop updating weights and biases values.




**Memorization:**

It is a trick in computer science in which when we have to use the output of one code repeatedly in future,that time we store it somewhere and use it directly without calculating it multiple times whenever we want the output.

By this trick,we reduce the time taken by code to give output but the space/memory required by code increases as we are saving the output of that function for future use. We use this same trick in backpropogation algorithm to save the time of training.

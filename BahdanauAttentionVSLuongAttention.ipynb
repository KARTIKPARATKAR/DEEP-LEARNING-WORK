**Attention Mechanism -**

Encoder Decoder architecture is not able to create the perfect context vector which passes the compressd input information to the decoder block which eventually gives output. So to resolve this issue we introduced with Attention Mechanism in which one or more hidden states of hidden LSTM time step is given as the input to the Hidden stste of LSTM cell of decoder so that decoder can focus his sttention to those cells only where important information is stored in a perticular hidden states.

Here we give the hidden state information of encoder LSTM to the hidden cells of decoder LSTM by drawing the full connected layer as above. Equation of input at first time step is c1 , equation of input at 2nd time step is c2 etc.In these equations , we know the hidden states h1,h2,h3...hn , but we dont know the values of alpha1,alpha2 ....alphan. These alpha values determines which input hidden states are important for calculating the ouput at decoder LSTM cell. SO there are 2 ways by which we can calculate the values of alpha and those are -

1)Bahdanau Attention

2)Luong Attention

**Bahdanau Attention**

Bahdanau Attention is also known as additive attention.It uses a feedforward neural network to calculate the values of alpha.It takes the weighted sum of encoder output.Previous decoder hidden state is used.This sttention is slightly more computationally expensive because we are introducing the feedforward neural network along with the LSTM models in encoder and decoder.

**Luong Attention-**

Luong Attention is also known as Multiplicative attention.It uses dot product to calculate the value of alpha.It uses current decoder hidden state which gives the more recent information about the input data.This attention is more computationally efficient because of dot product and absence of neural network.

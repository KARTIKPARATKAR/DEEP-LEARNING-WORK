**Forward Propogation-**

We have a very complex neural network architecture with multiple input variables,multiple hidden layers with multiple nodes and multiple or single output nodes having different activation functions like Sigmoid,ReLU,Softmax,Step,Linear and Tanh etc.

Forward Propogation involves the transmission of input data through the networks layers to produce an output.Each layer processes the input using weights and activation functions,ultimately generating a prediction.

**Forward Propogation Steps:**

1) Input Layer:

  X is a input feature. If we have 3 input features then X=[x1,x2,x3]

2)Compute Weighted Sum(Z):

  Z=W.X + b

  where b is a bias term

3)Apply Activation FUnction:

 To introduce non-linearity,an activation function like ReLU,Sigmoid or Softmax is applied:
    
     A=f(Z)

4) Pass to the next layer:

  The activated values become inputs for the next layer,repeating step2 and step3 until reaching the output layer.

   Z2 = W2A1 + b2

   A2 == f(Z2)

5)Compute Final Output:

 At the final layer,activation functions like SOftmax(for multiclass) or Sigmoid(For Binary classification) provide the final predictions

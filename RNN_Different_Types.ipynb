**Different Types Of RNN-**

Depending on the type of RNN , backpropagation algorithm in RNN vary.
There are 4 types of RNN architectres -

1) Many to One

2) One to Many

3) Many to Many

4) onr to one

**Many to One Model Architecture of RNN**


This architecture expects sequence in the input like sentence , characters or time-series and gives us the non-sequence data like integer or number or scalers.

This type of RNN architecture is used in Sentiment analysis. We have a sentence as a input in sentiment analysis problem to tell whether it is positive or negative.Which means output is single either 0 or one.

Many to one is used in rating predictions of movie.Input for this problem is sentence or group of sentences and output would be stars between 1 to 5 which is rating of the movie.This is multiclass classification problem.

We call it as a many to one architecture because we are giving multile inputs and getting only one output.


**One To Many Architecture of RNN-**

We get non-sequential data like image and output is sequential like words , sentence etc.We have a RNN blocks where we provide the inputs , since the input is single , so we will provide input for only one time.At every time step , we get different outputs.We are calling it as one-to-many architecture because we are provinding input for only one time and getting output multiple time at every time step.

This architecture of RNN is used in Image Captioning probelm where we give  a single image as input and it gives textual representaion of that image.We give a image where boy is playing with ball. IN output we get the texual representation of this image as "A boy is playing with ball" , which means we gave single input and in return we got multiple output in form of words.



**Many To Many RNN Architecture -**

In input as well as in output , we have  a sequential data . There are 2 types of M-M RNN as -> Same length many to many where we have input sequence length is same as output sequence length and another one is variable length many to many where we have length of ouput sequence is not same as length of input sequence.

In Same length many to many RNN , we have a problem of knowing the parts of speech of the sentence. So we pass the sentence to RNN model and it will give the part os speech of each word in the sentence.Here input sequence length and output sequence length is same.

In variable length many to many RNN , machine translation where we have one sentence in English and want to translate it in marathi language. So the number of words in english will not be same to number of words in Marathi. SO the input sequence length is not same as output sequence length.elow is the many to many RNN model -

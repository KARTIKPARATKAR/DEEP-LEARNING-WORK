**Problem In ReLU Activation Function-**

The problem that occurs in Rectify Linear Unit is known as **Dying Relu Problem.**  Focus on formula of ReLU activation function, it is max(0,x) which means if the output of a neuron is negative then the ouput of activation funtion is 0. If ouput of activation function is 0 then it is dead neuron and wont contribute to the architecture of the neural network. We can consider this neuron as it is not present because its output is 0.Once the output of neuron becomes 0 then it will forever be 0 during the course of training.

And if this problem of output of neuron is 0 happens with more than 50% of the neurons present in the neural network then our model wont work perfectly because neural network wont be able to extract complex pattens from the data as most of the neurons are dead due to its 0 output. And this problem is known as **Dying Relu Problem**.




**Why Dying ReLU Problem Occurs In Neural Network-**

1.jpg

Consider a single neuron in neural network with 2 inputs x1 and x2 , weight values as w1 and w2 with bias b1. We are applying ReLU activation function to this neuron. Formula of ReLU is max(0,x).  Output of the neuron is z1 and it is calculated as weighted sum of weights and inputs and bias as -

z1 = w1x1 +w2x2+ b1

If this z1 value is negative then the output of ReLU activation function will be 0. And the derivative of this 0 will also be 0.i.e. del a1 / del z1 will be 0.

When we update the values of weight and biases to get minimum loss function values, we use formula as

w1 = w1 - n * del L / del w1

w2   = w2 - n * del L / del w2

where n is the learning rate.

To find del L / del w1 and del L / del w2 , we have the folowing chain rule.



1.jpg

HEre we observe that we have a term del a1 / del z1 term is chain rule but we have its value as 0. So in the backpropogation updation formula ,that second term will also become 0 and there will be no updation happening throughout the training of the neuron.

The problem is arising due to negative value of z1 . That is -

z1 = w1x1 + w2x2 + b1

The value of z1 will be negative in case if you consider higher learning rate and if you consider high negative bias.


**Solutions To Overcome Dying ReLU Problem-**

1)Set low learning rate

2)Set positive bias value and it is experimentally proven that set bias as 0.01 which probably wont give dying ReLU problem.

3) Dont use relu, instead of it use Variants of ReLU like Leaky ReLU , Parametric ReLU , Elu ,Sel etc.

**Two types of Variants of ReLU-**

1) Linear Relu Variants- Obtained when linear transformation is applied on ReLU

a) Leaky ReLU

b) Parametric ReLU

2) Non-Linear Variants- Obtained when non-linear transformation is applied on ReLU.

a) Elu - Exponential Linear Unit

b) Selu - Scaled Exponential Linear Unit

**Leaky ReLU-**

1.jpg

Leaky ReLU is linear variant of ReLu obtained when linear transformation is apllied on ReLU.

Formula of leaky is - f(z) = max(0.01z,z)

If value of z is positive then output of leaky relu is z and if the value of z is negative then output of leaky relu is z/100

This makes sure that we dont get output as 0 which solves the kind of dying relu problem by the presence of some value for negative graph.

Advantages- Leaky ReLU is non saturated activation function. It is easily computed as there is no exponential term in formula. There is no dying relu problem. Leaky ReLU is close to zero centered



**Prametric ReLU-**

1.jpg

In Leaky ReLU , we multiply the output by 0.01 so that there should be some output for negative values.But why to take only 0.01 and not any other value?

So if we replace that 0.01 value with 'a' which is a trainable parameter then this activation function becomes Parameteric ReLU. The formula of Parameteric becomes-

f(x) =| x , if x>0 and ax , otherwise

During the training of the neural network , the value of trainable prarmeter 'a' is found out on a given data.



**Elu - Exponential Linear Unit-**

1.jpg

Elu is same as Parametric ReLu only with the difference in exponential term in formula.For negative part of the graph, there is a exponential curve which in some cases performs better to capture non-linearity of the data.

Value of Elu is close to zero centered which results in fast convergence. It is experimentally proven that generalization result is better in Elu.There is no dying relu problem.Elu is always continuous and differentiable.

Elu is computationally expensive because there is exponential term in formula of Elu.

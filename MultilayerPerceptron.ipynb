**Need of Multilayer Perceptron(MLP) when we had Perceptron-**

Perceptron is nothing but a straight line whose equation is Ax + By + C =0 .
So when we use this perceptron for classification problem wherein dataset is not linearly separable, our perceptron model will give very low accuracy on classifying data in either of the class. To overcome this issue, we have a multilayer perceptron model which can get a better accuracy of classification problem when we have non-linear dataset.

1.jpg

Now we know Perceptron model will not work properly in classification problem of non-linearly separable data. But still we build two different perceptron model with different activation function. These two activation function will give a decision boundary as shown in above image.Let those two decision boundry be 2x+3y+6=0 and 5x+4y+3=0. What if we somehow combine these two models and get a combined decision boundry which will look like above image. This combined decision boundry kind of more accurate in classifying the non-linear dataset in two different decision boundry. This decision boundry will need some smoothening but still it is better than a single perceptron model.

Therefore in a way you are adding two perceptron which is a multilayer perceptron to achieve a better accuracy model which is combined decision boundry.
So this is nothing but a Multilayer Perceptron Model.

We will consider 3 percptron model and train all 3 models on more -non-linearly separable data.

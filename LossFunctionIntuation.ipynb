**Problem With Perceptron Rule-**

Equation of perceptron is - Ax+By+b=0

We use perceptron rule to find the best line equation i.e. perceptron equation to distinguish all the datapoints correctly in one of the region. We assign random values of w1,w2 and b and check the random datapoint if it is distinguished correctly or not, if not then we will change the values of w1,w2 and b so as to distinguish that point correctly in a region. It might happen like when we re selecting random pints from dataset, it is selecting only correctly distinguished points only which may lead us to wrong values of w1,w2 and b. And second problem is everytime we execute the perceptron rule we might get the different values of w1,w2 and b. So everytime we will get different equation of perceptron. We cant rely on the values that we got by executing percptron rule to find the values of w1,w2 and b . So perceptron rule is not the perfect solution to find the values of weights and bias

**Loss Function-**

Loss function measures how far an estimated value is from its true value.This can be considered as the error made by our model. We aim to have a minimum loss function value so as to get better prediction.

Lets consider we have different w1,w2 and b value combination and we want to select the best values that draws correct perceptron eqution which can distinguish datapoint correctly in one of the region. What we will do is, we will find the loss function value for different w1,w2 and b values and select the w1,w2 and b value whcih has the lowest loss function value.

Loss function is literally a mathematical function.

- First type of loss function might be the number of misclassified points , which means number of wrongly classified points. If no. of number of wrongly classified points are more then loss function value is more and if the number of wrongly classified points are less then loss function value is also less.

- Second type of loss function might be to calculate the sum of perpendicular distance between wrongly classified datapoints and the percptron equation line. If this distance is more then loss function is also more and if this distance is less then the loss function value will also be less.This type of loss function is more accurate as it gives kind of higher weight to the datapoints which is being wrongly classified as its distnace between percptron line and datapoint will be high.

- Third type of loss function might be to put the coordinate values of wrongly classified datapoints in a percptron equation and taking sum of all the values of all wrongly classified datapoints. If this value is higher then loss function value is also higher and if this value is lowe then the loss function value is also lower.

**Loss Function-**

It is a method of evaluating how well your algorithm is modelling your dataset.

If loss function value is very high then our model is performing very bad and if the output of loss function is low then our model is performing better.

Loss function is nothing but a mathematical function of parameters of neural network.

**Why loss function is importnant-**

There is a saying of "You cant improve what you cant measure". So loss function measures the error made by our model.So after measuring this error we get to know that how can we improve the performance of model. So for improving the performance of model , loss functions play important role.

**1)Mean Squared Error-**

1)MSE is commonly used loss function in deep learning.

2)MSE = (actual value - predicted value)**2

3)We are taking square of difference because if the difference is more then we need to change the value of parameter more, so why not square this difference so as to get the higher values so that our model will change the values more so as to achive a better parameter values.

 1 square is 1

 2 square is 4

 3 square is 9

 So if error is increasing then our punishment to the model will also be high.

4) MSE is easy to interpret. MSE is differentiable so when we apply Gradient Descent technique , differnetiation is easy. MSE has only one local minimum which we are aiming.

5) Calculated MSE will have the unit in square because we are squaring the value.MSE is not robust to outliers.

6) You can only use MSE loss function when the activation function of output layer is linear function.


**2)Mean Absolute Error-**

1)MAE = |output value - actual value|

2)MAE is very easy to understand. Unit of MAE-loss function is same as the output value because we are not squaring it. It is robust to outliers.

3)MAE is not differentiable so when you apply Gradient Descent technique which is heavily based on differentiation,it doent work easily.


**3)Huber Loss**

1) Huber loss is a combination of MSE and MAE , when you have have 25% data as like outlier. So we cant treat those 25% points as outlier. In this case neither MSE nor MAE will give us a better loss function, in this case we will use Huber Loss.

**4)Binary Cross Entropy-(BCE)**

1)BCE loss function is used in binary classification problem which has only two classes to predict eg- Yes or No, Male or Female etc.

2)When you are using Binary Cross Entropy loss function then make sure to use Sigmoid Activation function in the output layer.

3)BCE is differentiable so we can use Gradient Descent easily. There might be multiple local minima in BCE. BCE is also known as log loss.

1)CCE is used in multiclass classification problem wherein we are predicting 3 or more classes. We have to use Softmax activation function for multiclass classification problem.

2)We define the equal number of output nodes as number of classes of classification. We will get the probability value at each of the output nodes. More the probability value , more is the chances of that perticular class to be predicted as output.

3)When we apply Categorical Cross Entropy loss function in multiclass classification problem with softmax as the activation function of output layer,we perform one-hot-encoding to the output columns.

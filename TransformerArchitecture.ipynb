**What is Transformers?**

Transformers are nothing but the neural network architecture like ANN,CNN,RNN,LSTM and GRU which is build to deal with sequence to sequence tasks like machine translation in which we translate the english text in hindi text.

In transformer also,we have encoder block and decoder block. We dont use LSTM cells in encoder and decoder blocks.Instead of it , we use self-attention which results in encoder to process the large text parallely.This parallel processing of training makes the training fast and scalable.And due to this we can train  this transformer architecture on large dataset.

Transformer architecture was introduced in research paper with title "Attention Is All You Need".

#**Origin Story Of Transformer-**

###1st research paper -

The origin story of transformer is introduced in 3 research papers.In 2014-15,research paper called **"Sequence to Sequence Learning with Neural Network"**.This paper states that we can use encoder-decoder architecture where LSTM cells are present in both encoder and decoder block to process the sequenctial data.Context vector passes the input compressed information to decoder block for final output. This architecture works better when we have words less than 30 words. As soon as number of words increase to more than 30 , it doesnt give bettter result because context vector can not pass the compressed information to decoder.


###2nd research paper -

To resolve the above problem , one another research paper was introduced called **"Neural Machine Translation By Jointly Learning To Align and Translate"**
This was the 1st paper who introduced the attention mechanism to resolve the issue of not transferring all the compressed input information to decoder block efficiently when we have more than 30 words in a sentence.Translation quality was improved when attention mechanism based encoder-decoder architecture was introduced.Though attention mechanism solved the problem of quality of translation in encoder-decoder architecture,there was still one problem existing of not able to train the encoder-decoder model parallely.As we have LSTM cells in encoder-decoder blocks , we had to train this architecture by sequence like passing each word of sentece in increasing time step.Due to this sequential training , we were not able to use the technique called Transfer Learning and we had to train the Encoder-Decoder model from scratch and we were not able to train the model  on huge data like Tera bytes in size.Training the model from scratch requires more time and money for labelling the data.


###3rd research paper -

To resolve the above problem of sequential training , one another research paper was introduced called **"Attention Is All You Need".** This paper introduced the self attention in encoder-decoder architecture.There was no LSTM/RNN layer in encoder and decoder blocks.This architecture can be trained parallely which results in high scalability of the model where we can train this architecture on huge dataset like Tera Bytes.This enabled the concept of transfer learning in Transformer Architecture.Based on this architecture only we got BERT(Bidirectional Encoder Representaation Of Transformer) and GPT(Generative Pre-Trained Transformer) which can be fine tuned and used for custome application in NLP.

Use of Self Attention instead of LSTM/RNN.

Architecture is stable.

Hyperparameters used in Self-Attention based encoder-decoder transformer architecture are very-very stable and robust.Robust means hypermparameter values used in this paper are still relevent in todays research and we can use same hyperparameter values to get state of the art results.

Research papers are generally incremental which means we take up one reseach paper and improve the different things in it and suggest the improved quality of the paper. But this paper is not kind of incremental which means this paper is so perfect that we can not suggest any increments in this paper.

**Gated Recurrent Unit (GRU)**


GRU is nothing but RNN architecure used to process sequential data.

Simple RNN cant understand the long term context due to vanishing gradient problem and exploding gradient problem.To resolve this problem in RNN , We introduced LSTM which can remember the long term context with short term memory and long term memory.

LSTM introduced in 1997 and GRU introduced in 2014. LSTM is having complex architecture and has lot of training parameters which makes the training of LSTM little slow. In GRU , it has kind of simpler architecture which result in lower number of trainable parameters and training of GRU is faster.Every other problem can be solved by LSTM and GRU and whichever gives better accuracy , we can make use of that.

**Big Idea Behind GRU**

Gated Recurrent Unit dont need two separate states to carry short term context and long term context.GRU is designed such that it is able to carry STM and LTM in single state.There are only 2 gates in GRU are Reset gate and Update Gate.

**The Setup of GRU**

For any time stamp t , we are giving two input , one is previous hidden state and second is the input of current time input.We are expecting the output of current time hidden state.

h is hidden state.

ht-1 is previous hidden state.

Xt is input.

rt is reset gate.

zt is update gate.

ht~ is candidate hidden state.

All these above are vectors.

Except input(xt) , all other has same dimentinality.

Yellow boxes are fully connected neural network layers.

xt is a vector becuase deep learning model can not understand the english like languages.So we need to convert it into numbers/vectors with vectorization technique.

**Architecture of Gated Recurrent Unit**

For any time stamp,inputs are previous hidden state and vectorized text and output is ht.There are 4 steps in calculating ht.

1) Calculation of rt (reset gate)

2) Calculating ht~ (candidate hidden state)

3) Calculating zt (Update Gate)

4) Calculation of ht (Curent Hidden State)

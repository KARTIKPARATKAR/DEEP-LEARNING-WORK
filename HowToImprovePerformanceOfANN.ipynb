**How to improve performance of Artificial Neural Network-**

**1)Fine tunning the neural network hyperparameters**.Hyperparameters means while defining an architecture of neural network,whatever we define by our own.
Hyperparameters are number of hidden layers,neurons per layer,learning rate,optimizer,batch size,activation function and number of epochs.

**2)By Solving the problems of Neural Network-**

a) Vanishing and Exploding gradient problem.

b) Not enough data

c)Slow training

d)Overfitting

**How to decide number of hidden layers:**

We have input , output and hidden layers in neural network. It is always preferred to take multiple hidden layers with limited neurons than the 1 hidden layer with lot of neurons.We have to experiment with number of hidden layer till we get the overfitting. As sson as we come across overfitting, we have to stop incresing number of hidden layers.

**How to decide number of neurons per layer in neural network:**

We are pretty sure that input layer will have the same number of neurons as number of features present in a dataset.Number of output neurons in output layer is 1 in case of regression problem, 2 in case of classification problem and if problem is multiclass classification then we have to define number of neurons at output layer as equal to the number of classes in output column.For number of neurons in hidden layer , we can start with sufficient number of neurons so that it can extract the features smoothly. If there is overfitting then accordingly decrese the number of neurons in hidden layers. You can also follow the pyramid structure wherein 1st hidden layer will have 64 neurons,2nd hidden layer will have 32 neurons and 3 rd hidden layer will have 16 neurons.But by experimentation it is found that following pyramid structure is not mandatory to increase the performance of a model.

**How to decide Batch Size?**

we have Batch Gradient Descent,Stochastic Gradient Descent and Mini-batch Stochastic Gradient.If you consider smaller batch size in range of 8 to 32 then learning will be slow and if you consider high batch size(8192) then learning will be fast.We get better performance on unseen data in case of smaller batch size.But if you could obtain a better results with high batch size then this will be considered better because it is fast learning.To get better result on high batch size we can use learning rate scheduler means when you are in initial epochs ,keep the learning rate very small and as your epochs increases , increase the learning rate which results in better performance of model with high batch size.This process of increasing learning rate is known as "Warning Up the Learning Rate".

So at the end, go with higher batch size,see if it is giving better performance or not. If it is not giving better performance then go with small batch size which is tried and tested and surely it will give us better results.

**How to decide Epoch value?**

We can set epochs as either 100,200 or 1000.Just make sure that you have applied early stopping which will detect the no-change in loss function value and it will stop the training as soon as we reach a convergence stage.

**Idea Of Pre-Trained Model-**

When we have a better models already trained on huge dataset with different classes of images , then if we solve any classification of image problem, we can make use of this pre-trained model so that time required to train the model decreases.

We have powerful models like VGGNet and ResNet trained of huge dataset of image like ImageNet,so we can make use of this model for our usecase.


If we use pre-trained models which means you dont have to train the model which means you dont need data or less data would also be enough to build the model.We can directly use the model without training.

There are different pre-trained models are available in Keras which is listed here - https://keras.io/api/applications/

**We will try to use ResNet50 pre-trained model in keras to predict what is there in the image ?**

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input,decode_predictions
import numpy as np
```

```python
#Now importing the ResNet50 model in model
model = ResNet50(weights='imagenet')
#Here we are defining that we want weights of parameters when model was trained on imagenet dataset.
```

```python
#Now we have to download and upload the image to predict what is there in the image
img_path = '/content/manonbeach.jpeg'
img = image.load_img(img_path,target_size=(224,224))
x = image.img_to_array(img)
x = np.expand_dims(x,axis=0)
x = preprocess_input(x)
```

```python
pred = model.predict(x)
#Here we are taking the prediction and storing it in pred variable
```

```python
print('Predicted:',decode_predictions(pred,top=3)[0])
#Printing top 3 predictions what ResNEt model thinks that image is
#If change 3 to 5 then it will give you top 5 predictions about your input image.
#To test the new image you have to give the path of image after uploading it to the files section of google collab
```

So this is how we built a universal classifier which gives the informtion regarding input image without training the model of our own . We used pretrained model of ResNet which is trained of ImageNet Dataset.

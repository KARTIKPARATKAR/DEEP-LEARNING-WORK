Here the multilayer perceptron model architecture is given. MLP model contains one input layer,one output layer and one or more hidden layer. x1 , x2 ....xn are input features to the model. y1,y2 ...ym are output of the model. Each connecting-layer with nodes will have its perticular weight which defines the importance of that feature in deciding the output.

Each node will have its bias value.For denoting the weight value of each connecting node we see from which input feature it is coming and to which node it is going to.

If connection is coming from feature x1 and going to the first hidden node then we will denote this weight as w11.

If connection is coming from feature x2 and going to the first hidden node then we will denote this weight as w21.

So in the above multilayer perceptron model architecture,we have one input and output layer each and 2 hidden layers. Number of connections that we have in an architecture will be the number of weight values that we have to find while training the MLP model.We can see 12 connections are connecting to the first hidden layer,6  connections are connecting to the second hidden layer and 2 connections are connecting to the output layer. So in total we have 12+6+2 = 20 weight values . We have 5 nodes in hidden layer and 1 node in output layer so in total we have 6 bias values. Each node will have its own bias value that we have to find while training the MLP model. Observe how we have given name to the weight values that connects nodes internally and bias values of each node.

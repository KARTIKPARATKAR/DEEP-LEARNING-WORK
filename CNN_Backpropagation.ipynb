**Backpropagation In CNN-**

Let we have a image of 6 * 6 pixel,which we are passing through the filter of 3 * 3 . There will be bias in filter. We will get the feature map after convolution layer of size 4 * 4. We will apply relu activation function of feature map.We will apply max-pooling layer of size 2 * 2 with stride of 2 resulting in the size of image in 2 * 2. Then we will apply flatten layer which gives us a 1D tensor and there will be 2 * 2 = 4 inputs which will be passed through the fully connected ANN and at the output layer we will get the output of 1 * 1 scaler.We get the trainable parameters at the convolution layer and fully connected ANN . There will be no trainable parameters at pooling layer and flatten layer.In total we will be having 9 trainable parameters of 3*3 filter with one bias value of filter. There are 4 inputs to to ANN so these 4 weight values along with the one bias value of node in ANN , so 9 + 1 + 4 + 1 = 15 trainable parameters present.

So whenaver we will train this convolutional neural network, we will have to find the optimum values of these 15 parameters so that the loss function is minimum.Here loss function will be binary cross entropy.

Our end goal is to apply gradient descnet on all the trainable parameters and find out the values of all parmeters to get the minimum loss.For understanding the backpropagation in complete CNN architecture , we can divide the model architecture in two parts , one is from input image to flatten layer and second will be from ouput of flatten layer to the final prediction.

Consider the second part of the model which has ANN network and apply backpropagation algorithm.

When we consider first part of the model which contains convolution layer , pooling layer and flatten layer. We have to apply backpropogation algorithm at all these three layers to know the best values of parameters so that we get minimum loss function.


For applying the gradient descent algorithm to find loss function at convolution layer. We have weight values of filters w1 and bias value b1 at this convolution layer as parameters whose optimum values we have to find out.Here w1 and b1 values are dependant on changes in z1 , z1 is dependant on A1, A1 is dependant on P1 and so on. So we will have to calculate all the values to find relation between change of value in w1 and b1 with repect to L(final output).

**Convex VS Non-Convex Optimization-**

Observe the following graphs which represents the convex vs non-convex optimization techniques.

1.jpg

There are 3 problems in non-convex optimization technique-

1)Local Minima- Optimization function can stuck in local minima which is not the desired solution of optimization problem.Local Minima is the suboptimum solution.SGD can not solve this local minima problem.

2)Saddle Point-It is a point where in one direction slope is increasing and in another direction slope is decreasing. SGD is not able to detect such points and can not achieve required optimum solution.

3) High Curvature - If radius is more then curvature is low and if radius is less then curvature is high.It is difficult to traverse in a high curvature and we find very high curvatures in non-optimization problem.

1.jpg

So Gradient Descent optimization is not capable of get rid of these three problems and that is the reason why we had to come with some advanced Gradient Descent techniques to fasten the training process and to achive the optimim solution.

** Momentum Optimization -**

When we deal with neural network,loss function of some complex neural network are not the simple curve i.e. not a simple convex curve. We have kind of complex curve of loss function. To reach at minimum loss function of such complex curve by simple Gradient Descent Optimization Function is not easy due to

1)High Curvature  

2)Inconsistant Gradient

3)Local minima problem.

And this is the reason why we use Momentum Gradient Descent or Momentum Optimization Technique which solves all the above mentioned three problems.

**How Momentum Optimization Works?**

Basically the whole idea is to reach to the minimum loss function value in the graph with correct parameter values. When we calculate gradients and if the gradient is directing us to go towards perticular direction to find the minimum/optimized solution , then we will increase the speed of function to go towards that direction. If the gradient is directiong us not to go towards perticular direction, then we will decrease the speed of function to go towards that direction. So as there is speed involved , we name it as a Momentum Optimization.

1.jpg


Observe the path to reach the optimum solution in Stochastic Gradient Descent and Stochastic Gradient Descent with Momentum. In case of momentum , there is no zig-zag motion of function and it is reaching faster to the optimum solution. In case of without momentum, there is a zig-zag motion of function and it is taking more time in comparison to SGD with momentum due to its zig-zag motion.


**Disadvantage of Momentum SGD -**

As there is a momentum in function for reaching the optimum solution in the curve,sometime it might come out of the optimum solution due to its momentum and oscillates in the curve to settle doen to the desired solution. So Momentum itself is the drawback of Momentum Stochastic gradient Descent optimization technique.

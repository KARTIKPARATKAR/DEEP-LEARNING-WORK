**GradientDescent -**

It is an optimization algorithm and it is the most common way to optimize neural networks.

Gradient Descent is a way to minimize loss function.

Consider the following Backpropogation algorithm,the overall process of selecting each row and updating the weights and biases after calculating the loss function is nothing but a GradientDescent.

1.jpg

There are three types of gradient descent which differe from each other on the basis of how much data we use to compute the gradient of the objective function.

* Batch - Uses entire dataset

* Stochastic - Uses single datapoint

* Mini-Batch - Uses small subset of the total data. i.e. mini-batches

**Batch Gradient Descent Algorithm-**

It is also known as Vanilla Gradient Descent. We consider entire dataset and then we make an update in weight and bias values.

Let we have epoch value as 10. And if we have 50 datapoints in total , so we will use all the 50 datapoints for calculating the loss function and number of updates happening in weight and bias value will be same as number of epochs.

**Stochastic Gradient Descent-**

Let we have 10 epochs. And if we have 50 datapoints in total. We first shuffle the entire 50 datapoints to avoid any kind of bias. In every epoch,weight and biases will be updated for number of datapoints times.So in each 10 epochs,weight and bias values will be updated 50 times. IN total updation will happen for 10*50 = 500 times.

Frequency of weight updation is higher than the Batch Gradient Descent.

* Which is faster Gradient Descent?

* Ans- Batch Gradient Descent is faster than Stochastic Gradient Descent because if you set batch size equal to the number of rows then it will be a batch stochastic gradient and it will take less time because the updation of weight and biases will be number of datapoints times only.If you set batch size equal to 1 then it will be a stochastic gradient descent and it will take more time because the updation of weight and biases will be happenting (no. of rows * no. of epochs) times.

* For a given number of epochs,which will work faster to converge?

* Ans - Stochastic Gradient Descent will work faster to converge because the number of times process of updation of weight and biases are more in Stochastic GD than Batch GD.We wont need more number of epochs in Stochastic GD to achieve convergence. We will require more number of epochs in Batch GD to achieve convergence.

Stochastic Gradient Descent will help the algorithm to move out of local minima as it tries to reach minimum loss function in very weaverd path. But in case of Batch Gradient Descent, if you stuck in local minima,probabily you wont be able to move out of local minima because Batch GD tries to achieve minimum loss function in kind of straight path.

**Mini-Batch Gradient Descent-**

This mini-batch GD is best of both Stochastic GD and Batch GD.Let we have 320 rows of data. If we create a 10 batches then each batch will contain total of 32 datapoints.Here for every epoch,we will update the values of weights and biases for 10 times.10 datapoints are taken into consideration for updating the values of weights and biases and we will do this process for number of epochs times.

Let we have 1000 rows in a dataset.

In case of Batch GD , all 1000 samples are used to compute the gradient.Batch size in Batch SG is equal to the number of rows in a dataset.

In case of Stochastic GD , it uses only one sample at a time to compute the gradient.Parameters are updated 1000 times  per epoch.Here dataset size is 1000 rows and batch size is 1.

In case of Mini-Batch Gradient Descent,it uses a small subset(batch) of the dataset to compute the gradient.Parameters are updated multiple times per epoch.Dataset size is 1000 rows and batch size 100 so number of updates per epoch is 1000/100 = 10

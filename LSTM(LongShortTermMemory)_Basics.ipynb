**LSTM - Long Short Term Memory-**

It is widely used in industrhy. The foundation step of LLM(Large Language Model) is LSTM. RNN dont work well when we have a lot of texual data because RNN dont remember the long term dependencies.LSTM was introduced due to RNN was unable to work better on big texual data.

**LSTM core Idea-**

LSTM works exactly as our mind remembers the long story.If you think , our mind identifies the long term important things nLomg Term Memory) and short term important things (Short Term Memory) and keeps the communication between these two memories. If mind decides that this perticular thing should go to long term memory then short term memory push the information to long term memory.

In RNN , we have only one memory which is only capable of remembering short term dependencies. But in LSTM , we have one Long Term Memory line where we keep the long term dependencies whenever we thought it is importance. These long term dependencies are pushed from short term dependencies and unless you remove the long term dependency , it will go to the end which is to the output.




**Difference Between RNN and LSTM**

In **RNN** , we have a simple architecture inside the node and have only one memory remembering system which is state which is nothing but a short term memory. This state of the RNN is not capable of remembering long term dependencies.

In **LSTM** , we have two memory system , one is Short Term Memory and another is Long Term Memory . STM keeps the short term dependencies and LTM keeps the long term dependencies pushed by STM whenever we felt like this is a long term dependency. To have a communication between short term dependency and long term dependency , we require a complex type of architecture.

**There are 3 Gates in LSTM -**

1)Forget Gate - Based on the current input and the short term context, It decides what things to remove from Long Term Memory.

2)Input Gate - Based on current input , It decides what information should be added to the Long Term Memory.

3)Output Gate - Based on current input , It decides what should be taken from long term memory and show as a output.At every given stage,output gate create a short term memory.

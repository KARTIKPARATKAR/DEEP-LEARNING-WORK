**What are activation function in Neural Network?**

1.jpg

Activation function is the mathematical gate between input and output of the neuron.Activation function only decides whether that node will get activated or not and if it is getting activated then how much wil it get activate is also decided by activation function.

If we dont include activation function in neural network then our neural network wont be able to capture the non-linear data , it will capture only linear data. If we dont use activation function then our model will work only as linear regression and it wont work on non-linear data.Activation function introduces non-linearity in the neural network.

Now we will consider a non-linear separable data and try to build a neural network with linear activation function and see whether it is able to classify the data in two regions properly or not.

```python
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
#Generating 2d non linear data
x,y = make_moons(n_samples=500,noise =0.2,random_state = 2)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap="coolwarm",edgecolors='k')
plt.title("Non-Linearly Separable Data")
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

While defining the sequential model , we gave activation function as linear so you can consider that we havent given any activation function to our neural network

```python
#Now defining the model
model = Sequential()
model.add(Dense(128,input_dim=2,activation = "linear"))
model.add(Dense(128,activation="linear"))
model.add(Dense(1,activation='sigmoid'))
adam = Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
history = model.fit(x,y,epochs=500,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x,y.astype('int'),clf=model,legend=2)
plt.xlim(-0.7,0.5)
plt.ylim(-0.8,0.8)
plt.show()
```

SO above is the decision boundry that is drawn which is not correctly classifying the square and triangle datapoints in data because our data is not linear separable.We rewuire the model which has a curve like structure so that it can classify the non-linearly separable data with good accuracy.

SO we can say that with the linear activation function,our neural network model will work as a linear regression or logistic regression model only.

Now we will define another model with activation function at hidden layers as relu,lets see what changes we can observe.

```python
model2 = Sequential()
model2.add(Dense(128,input_dim=2,activation = "relu"))
model2.add(Dense(128,activation="relu"))
model2.add(Dense(1,activation='sigmoid'))
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
history = model.fit(x,y,epochs=500,validation_split=0.2)
```

```python
plot_decision_regions(x,y.astype('int'),clf=model2,legend=2)
plt.xlim(-0.7,0.5)
plt.ylim(-0.8,0.8)
plt.show()
```

Here observe that our model is better predicting the two different classes of ouput data with L shaped model. This non-linearity is introduced to the model due to use of relu activation function in both the hidden layer.

**What is the ideal activation function quality?**

1)Activation function should be non-linear to capture the non-linearity in the dataset

2)Activation function should be differentiable because when we use gradient descent, it differentiates the activation function and updates the weights and biases.

3) Activation function should be computationally inexpensive. Calculating the derivatives should be simple,easy and fast otherwise training of the model will be slow.

4)Activation function should be such that it should be zero centered or it should be normalized.tanh is the the best example of zero centered activation function.

5) Activation function should be non saturating function.ReLu is non-saturating function because it doesnt sqeeze the output in any perticular range like sigmoid ranges from 0 to 1 and tanh ranges from -1 to 1.
Vanishing Gradient Problem arises in saturating activation function because it gives output in perticular range and when we do backpropgation and update the weights,its value is very small so the updation in weights and bias values will also be very small or in some case it doent get updated. This problem of non-updation is known as Vanishing Gradient Probelm.

**SIgmoid Activation Function:**

1.jpg

Advantages of Sigmoid Activation Function-

1) As output of sigmoid activation function ranges from [0,1] , so we can treat the output as a probability.We use sigmoid activation function in binary classification problem.

2)Sigmoid is a non-linear activation function so it is capable of non-linear data.

3)Sigmoid is differentiable at every point of time.

Disadvantages of Sigmoid Activation Function-

1)Sigmoid is a saturating activation function,so Vanishing Gradient Problem arises when we use Sigmoid Activation Function.

2)Sigmoid activation function is non-zero centered which results in slow training of the model which ultimateky effect on convergence of the model.

3)Sigmoid activation function has complex formula which includes e terms and minus power term. So this function is computationally expensive so for convergence of the model , it takes more time.

**Tanh Activation Function(Hyperbolic Tangenet)-**

1.jpg

1)Range of tanh is -1 to 1 where sigmoid ranges from 0 to 1

2)Formula for tanh is 2 times sig(2x) - 1

3) Tanh is non- linear along with differentiable and it is 0 centered which results in when we calculate its gradients,they are some +ve and -ve so the training is faster in comparison to sigmoid.

4)Tanh is saturating function which means it transforms output in a perticular range. So for bigger values , the slope becomes 0 and it leads to Vanishing Gradient Problem.

5)Tanh is computationally expensive because we have e power values in formula.

**ReLU Activation Function-**

1.jpg

0)ReLU is the most used activation function in deep learning.

1)Formula of ReLU is max(0,x) which means it gives same value if input is positive and if input is negative then it gives output as 0.

2)ReLU is non linear activation function. ReLU is not saturated in positive region because it can take any value in positive region because its range is from 0 to infinity.ReLU is computationally inexpensive because we have to derivative of only x. Convergence of ReLU is faster than the sigmoid and tanh activation function.

3)Disadvantages- ReLU is not completely differentiable.To say more specific, ReLU is not differentiable at x=0.ReLU is not zero centered.We use batch normaliztion technique to overcome the problem of non-zero centered.

**Activation Function:**

It is a mathematical function applied to the output of a neuron.It introduces non-linearity into the model,allowing the network to learn and represent complex patterns in the data. Without activation function,neural-network will work like a simple linear model.

**1)Sigmoid Activation Function-**


It is defined as 1 upon 1 plus e power minus x. The output ranges between 0 and 1 hence it is useful for binary classification. It shifts the output in the range of 0 to 1.

**2)Step Function(Binary Activatin)-**

If the output of neuron is greater than threshold then output of step function will be 1 and if output of neuron is less than 0 then output of step function is 0.

1.jpg


**3)Tanh Function(Hyperbolic Tangent)-**

Tanh is a shifted version of the sigmoid functions allowing it to stretch across the y-axis. Output values ranges from -1 to 1. Tanh activation function is basically used in hidden layyers due to its zero-centered output.

**tanh(x) = 2 times sigmoid(2x) -1**

**4)ReLU(Rectified Linear Unit)-**

 ReLu is the rectified linear unit. It is the most widely used activation function.

 **f(x) = max(0,x)**

The range of ReLU function is from 0 to infinity. ReLu is the default activation functio in deep learning.

**5)Softmax Function-**

Softmax is a activation function used in multi-class classification problem.It gives probability of each class which we can call it as a probability distribution.Higher the probability value , higher is the classification of datapoint in that class.Sum of all the probability values for all the classes is 1.

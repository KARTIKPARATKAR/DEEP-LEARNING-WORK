**Adam Optimizer-**

It is an Adaptive Moment Estimation optimization technique.Adam is the most popular optimization technique in Artificial Neural Network,Convolutional Neural Network and Recurrecnt Neural Network.

Till Now we have learned Stochastic Gradient Descent,Batch Gradient Descent, Mini-Batch Gradient Descent , Momentum GD , Nesterov Accelerated Gradient , Adagrad .

Adam merges the concept of momentum and learning decay which was used by previous optimizers to enhance the speed of reaching to the optimum solution of minimum loss function value.

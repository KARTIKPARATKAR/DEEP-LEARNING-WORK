Perceptron is an algorithm which is used for supervised machine learning. Perceptron is a building block of deep learning. For understanding, we can call Perceptron as a mathematical model or function.

**Now we will try to draw a perceptron which distinguishes the student is placed or not and we will visualize the output perceptron line.**

Dataset used is - https://github.com/KARTIKPARATKAR/DEEP-LEARNING-WORK/blob/main/placement.csv


```python
#Importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the dataset in pandas dataframe
df = pd.read_csv('/content/placement.csv')
```

```python
#Printing first 5 rows of the dataset
df.head()
```

```python
print(df.shape) #Dataset has 100 rows and 3 columns
```

```python
#Visualize the dataset
sns.scatterplot(x='cgpa',y='resume_score',hue='placed',data=df)
plt.xlabel('cgpa')
plt.ylabel('resume_score')
plt.title('Scatter Plot of cgpa VS resume_score')
plt.show()
```

So we can clearly see that this dataset is linearly separable by the straight line equation which is nothing but a Perceptron

```python
#Importing perceptron from sklearn library
from sklearn.linear_model import Perceptron
perceptron_model=Perceptron() #Here we are defining the perceptron model.
```

```python
#Selecting the data as input and output in x and y variable
x=df[['cgpa','resume_score']]
y=df['placed']
```

```python
#This .fit function will trains the perceptronby adjusting its
#weights based on the given data
perceptron_model.fit(x,y)
```

```python
perceptron_model.coef_
#This step will give the output values of weight .
#We got w1 value as 40.26 and w2 value as -36
```

```python
perceptron_model.intercept_
#This intercept function will give the bias value to be added in a equation
#We got -25 as the bias value to be added in equation of percptron
```

```python
#We want to see the line which distinguishes the dataset in two regions
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.values,y.values,clf=perceptron_model,legend=2)

#x.values and y.values are converting from pandas dataframe to numpy array
#clf = perceptron_model means trained perceptron model is used for classification
#legend=2 means it displays the legend at a specific location
```

So here we can observe the line which kind of distinguish the dataset in two different region of two different color.

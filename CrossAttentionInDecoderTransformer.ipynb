**What is Cross-Attention?**

Cross Attention is a mechanism used in transformer architectures,perticularly in tasks involving sequence to sequence data like translation or summarization. It allows a model to focus on different parts of an input sequence when generating an output sequence.

Let we have a translation problem of English to Hindi conversion.We have English sentence as "I like eating ice-cream".We give this sentence as input to the Encoder block of the model.Encoder understands the sentence and creates the compressed information and pass this information to the decoder.Decoder takes this compressed information as input and generates the output sentence in Autoregressive mode.

Let we are at time step 3 of decoder and we want to generate the 3rd word. What would it be? On what things do prediction of next word is dependant upon? It depends on what decoder has generated till now in 1st and 2nd time step and what is there written in input sentence? Self_attention tell what the generator has generated till now.Cross_attetion draws the similarity in two different sequence. We have two different sequence of "I like eating ice-cream" and "Mujhe icecream khana pasand hai". And we draws corelation of one word with all other words in another sequence.

**Self_attention VS Cross_Attention-**

Cross attention is conceptually very similar to self attention.

**The Input-**

In self_Attention,we send input as contextual embedding of previous self attention layer.

In cross attention , we send input as embedding or contextual embeddings of input sentence along with the embeddings or contextual embeddings of translation of the input sentence.

1.jpg

**The Processing inside the Self Attention and Cross Attention-**

In self attention,we create 3 vectors for value , key and vector part which plays different roles in calculation of contextual word embeddings of each word.

1.jpg

1.jpg

In cross attention, the difference is in the creation of weight matrix which is going to be used for calculation of vector , value and key vector. For every hindi word we have 1 query vector and for every hindi word , we have one key vector and one value vector.

1.jpg

1.jpg

**The output of self_attention and Cross_Attention-**

In self_attention , we send embedding of each word to the self_attention block and gets the ouput as contextual word embeddings of each word.

In cross_attention , we send word embeddings of each word in input sentence and word embeddings of each word in translated sentence and gets the output of contextual word embeddings of the translated sentence.

1.jpg

Cross Attention is kind of similar to the Bahadanau Attention in simple encoder-decoder architecture.

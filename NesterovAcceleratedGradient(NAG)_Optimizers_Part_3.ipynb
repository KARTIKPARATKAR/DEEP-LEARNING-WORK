**Nesterov Accelerated Gradient (NAG)-**

NAG is the advanced version of Momentum Stochastic Gradient Descent in which NAG solves the problem of Momentum where the path were having spirat route,oscillatory route and zig - zag route. With the use of NAG , you prevent such spiral and zig-zag path to reach to the optimum solution and as a result the training speed also increases.
So NAG has kind of simpler route by decreasing the number of spiral route or oscillations and zig-zag route and due to less zig - zag and spiral route , you reach to the optimum solution faster than the Momentum Stochastic Gradient Descent.

In case of Momentum SGD , we calculate momentum and current gradient at a perticular point and apply it , due to the application of both gradient and momentum at the same time it takes kind of longer jump and as a result it has oscillatory , zig-zag or spiral path to reach to the minimum value of loss function. In case of Nesterov Accelerated Gradient , we first calculate look ahead point with only application of gradient and make the movement toward wherever it is suggesting to go.Here we are making the movement by only gradient so this movement is kind od lesser than the momentum SGD which results in the simpler path to reach to the minimum value of loss function of the model.

**Disadvantages of NAG-**

Basically NAG makes the path of reaching towards minimum loss function simpler by damping the oscllation or decreasing the zig-zag path etc. This might result in the function to stuck in local minima.

**CNN VS ANN**

Broadly CNN is used when we have a image data.If we use image data on CNN we face issues like computational cost,overfitting and loss of important features like spatial arrangement of pixels.


Let we have a mnist data which has images of 28 * 28 pixels where digits are written. If you want to define ANN for this problem , we will have to convert this pixels into 1D i.e. 28 * 28 = 784 inputs. So we will have 784 inputs to the neural network.

If we want to define CNN for this digits classification problems , then CNN use this 2D image in 2D format only.We have filters which gives output as feature map.If filter is of 3 * 3 then we will get feature map of 26 * 26 .We add bias to each value in feature map.We apply activation function to the feature map.Then apply flatten layer to the output of activated feature map.Then we apply max pooling layer and flatten layer. After this we apply fully connected layer and lastly we have softmax layer where we get the output.

**How CNN and ANN are kind of same-**

If you bradly see , CNN and ANN are kind of similar.In ANN we have 784 inputs and we are passing to the next layer. In CNN we have 2D image where we are performing convolution operation which gives feature map. We are adding bias to this feature map and passing through the ReLU activation function.

So this is what kind of process we do in ANN also, We multiply w1x1+w2x2+...+w784x784 , then add bias to this weighted sum (b) and then apply activation to the resultant. We can understand adding a filter in CNN is kind of similar to adding a node in ANN.

**Question-** How many trainable parameters are there in CNN where image of 28 * 28 * 3 is used as input with 50 filters of size 3 * 3 * 3 ?

**Answer-** We will get the feature map of convolution process of the size of 226 * 226 * 50 as the fiter size is 3 * 3 * 3 . As filter size is 3 * 3 * 3 so number of weights are 27. We are using 50 filters so we will have total of 27 * 50 = 1350 weight values. We will be having a bias values of each filter which is learnable parameter, so 50 biases. In total we will have 1350 + 50 = 1400 learnable parameters in this network.

Main thing to notice here is that number  of learnable parameter is independant of the size of image that we are using. If you use 1080 * 1080 * 3 size of image, then also the number of learnable parameters will be same as 1400 because number of learnable parameters are independant of the size of input data.
